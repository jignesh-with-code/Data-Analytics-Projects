### Age
**Conclusion**
* not a normal distribution.
  Age is skewness for right side.
* maximun sales Age is 18 to 22 percent people and Age 23 to 25 his second high sales and then Age 45 to 50 third position sales and other is low.
* null values 173 then fills/remove the empty values.
* 0.6 percent data missing.
* feature engennering make a Age Based Age_Group Column.
* not any outliers in Age column.
* total count values different then other columns this is proble back check it.

```python
# apply the Descriptive Saticstics in Age columns.
# check the max/min/IQR of data.
df['Age'].describe()
```

```python
# Analysis the Chart of Age columns.
# i know that what is skewness of the data.
counts, bins, patches = plt.hist(df['Age'] , bins=15)
plt.xlabel('Age')
plt.ylabel('Frequency')
for count, patch in zip(counts, patches):
    plt.text(patch.get_x() + patch.get_width() / 2, count, int(count),
             ha='center', va='bottom')
plt.show()
```

```python
# kde(Kernel Density Estimate) check the Age column is normal distribution or notb.
df['Age'].plot(kind='kde')
```

```python
# cheking the skewess of the Age column.
df['Age'].skew()
```

```python
# check the outliers of the data in data set using box-plot.
df['Age'].plot(kind='box')
```

```python
#cheking the null values in Age columns.

Age_null = round(df['Age'].isnull().sum()/len(df['Age'])*100,2)
print(Age_null)


```
